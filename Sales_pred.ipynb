```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sns
```

```python
data=pd.read_csv("F:\\Downloads\\tvmarketing.csv")
print(data)
```

```python
print(data.info())# we can get the information about data like data type memory null value
```

```python
X=data.iloc[:,0:1].values #extraction features like x as a array all rows and 0 column return 2d array
y=data.iloc[:,1].values #extracting target which is output retun 1D array values which is convert to array 
print("X features\n",X)
print("Y Output:\n ",y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test,=train_test_split(X,y,test_size=0.3, random_state=42)
print("X_train\n",X_train)
print("X_test\n",X_test)
print("y_train\n",y_train)
print("y_test\n",y_test)

```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X_train,y_train)
```

```python
print("Interception: ",lin_reg.intercept_)
print("Slop: ",lin_reg.coef_[0])
```

```python
def viz():
    plt.scatter(X_test,y_test,color="green")
    plt.plot(X_train,lin_reg.predict(X_train),color="blue")
    plt.title("Linear Regression Model")
    plt.xlabel("TV")
    plt.ylabel("Sales")
    plt.show()
    
    
```

```python
viz()
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
y_train_pred=lin_reg.predict(X_train)
print(y_train_pred)
```

```python
y_test_pred=lin_reg.predict(X_test)
print(y_test_pred)
```

```python
mse_train=mean_squared_error(y_train,y_train_pred)
mse_test=mean_squared_error(y_test,y_test_pred)
r2_train=r2_score(y_train,y_train_pred)
r2_test=r2_score(y_test,y_test_pred)
print("MSE",mse_train)
print("MSE",mse_test)
print("r2_Square",r2_train)
print("r2_Suare",r2_test)
```
